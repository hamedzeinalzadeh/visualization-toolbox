# KDE(Kernel Density Estimation)plot
---

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna()
cars.shape
```

```python
cars.head()
```

## Univariate(1D) KDEplot

```python
sns.kdeplot(cars['cylinders'])
```

```python
sns.kdeplot(cars['cylinders'], fill=True)
```

```python
sns.kdeplot(cars['cylinders'], bw_method=5)
```

```python
sns.kdeplot(cars['cylinders'], bw_method=0.05)
```

```python
# For cumulative plot, scipy must be imported
import scipy
sns.kdeplot(cars['cylinders'], cumulative=True)
```

## Bivariate (2D) KDEplot

```python
sns.set_style('darkgrid')
```

```python
sns.scatterplot(x=cars.horsepower, y=cars.mpg, color='green', alpha=0.5, s=20)
plt.xlabel('Horse Power', fontsize=10)
plt.ylabel('MPG', fontsize=10)
plt.tight_layout()
```

```python
sns.kdeplot(x=cars.horsepower, y=cars.mpg)
sns.scatterplot(x=cars.horsepower, y=cars.mpg, color='green', alpha=0.5, s=20, zorder=10)
plt.xlabel('Horse Power', fontsize=10)
plt.ylabel('MPG', fontsize=10)
plt.tight_layout()
```

```python
sns.kdeplot(x=cars.horsepower, y=cars.mpg, n_levels=20)
```

```python
sns.kdeplot(x=cars.horsepower, y=cars.mpg, n_levels=10, fill=True)
```

```python
sns.kdeplot(x=cars.horsepower, y=cars.mpg, n_levels=10, fill=True, thresh=0.7)
```

```python
sns.set_style('whitegrid')
sns.kdeplot(x=cars.horsepower, y=cars.mpg,
            fill=True,
            cbar=True)
```

## 2D KDEplot by Category

```python
sns.set_style('darkgrid')

cyl_5 = cars[cars.cylinders == 5]
cyl_8 = cars[cars.cylinders == 8]

plt.figure(figsize=(8, 6))

sns.kdeplot(x=cyl_5.horsepower, y=cyl_5.mpg,
            cmap='Blues', fill=True, thresh=0.1, cbar=True)
sns.kdeplot(x=cyl_8.horsepower, y=cyl_8.mpg,
            cmap='Reds', fill=True, thresh=0.4, cbar=True)

plt.xlabel('Horse Power', fontsize=10, fontweight='bold')
plt.ylabel('MPG', fontsize=10, fontweight='bold')

plt.annotate('5 Cylinders', (100, 32), color='blue', fontsize=12, fontweight='bold')
plt.annotate('8 Cylinders', (180, 18), color='red', fontsize=12, fontweight='bold')
```

## 2D KDEplot + Distplot

```python
sns.histplot(cars.horsepower, kde=True)
```

## 2D KDEplot + Pairplot

```python
sns.pairplot(cars[cars.model_year.isin([80,81, 82])],
             hue='origin',
             vars=['mpg', 'horsepower', 'cylinders'],
             palette='colorblind'
            )
```
