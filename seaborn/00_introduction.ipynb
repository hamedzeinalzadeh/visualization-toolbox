```python
plt.plot(x, y)
```

## Seaborn + Pandas
**Seaborn** can group, aggregate, and plot **Pandas** `dataframe`.

```python
cars = sns.load_dataset('mpg')
```

```python
type(cars)
```

```python
cars.shape
```

```python
cars.dropna(inplace=True)
cars.shape
```

```python
cars.head()
```

```python
sns.relplot(data=cars, x='model_year', y='mpg', col='origin', hue='cylinders')
```
